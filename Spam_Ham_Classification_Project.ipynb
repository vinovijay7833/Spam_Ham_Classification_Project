```python
import pandas as pd
ds=pd.read_csv("/content/SMSSpamCollection.txt",sep="\t",names=["label","message"])
```

#Text cleaning and preprocessing


```python
!pip install nltk
```

```python
import nltk
import re
```

```python
from nltk.tokenize import wordpunct_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words("english"))
```

```python
Corpus=[]
for i in range(0,len(ds)):
  review=re.sub("[^a-zA-z]"," ",ds["message"][i])
  review=review.lower()
  review=review.split()
  stemmed_words=[WordNetLemmatizer().lemmatize(word) for word in review if word not in stop_words]
  review=" ".join(review)
  Corpus.append(review)
```

```python
Corpus
```

#Bag of words (Word to Vectorization)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,ngram_range=(1,2))
X=cv.fit_transform(Corpus).toarray()
```

```python
##Indepentant feature
X
```

```python
##Dependent feature
Y=pd.get_dummies(ds["label"])
```

```python
Y=Y.astype(int)
Y
```

```python
Y=Y.iloc[:,1].values
```

```python
X
```

```python
Y
```

##Machine Learning Model Training

```python
##Splitting the datasets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20)
```

```python
##Model selection
from sklearn.naive_bayes import MultinomialNB
Model=MultinomialNB()
```

```python
Model=Model.fit(X_train,Y_train)
```

```python
Y_P=Model.predict(X_test)
```

##Model Performance Analysing

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
Accuracy_score=accuracy_score(Y_test,Y_P)
```

```python
Accuracy_score
```

```python
Classification_report=classification_report(Y_test,Y_P)
```
